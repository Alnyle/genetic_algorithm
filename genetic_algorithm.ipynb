### Build the algorithm 
after finshing setting our parameter now let's build our model 


#### Model paratemer:
- `pop_size`: Pupulation size
- `mut_rate`: Mutation rate
- `radnom_state`: In case user want to start with same population
- `ppl`: Population
- `genes`: Options from which our population would be created
- `Target`: Target solution

```python
class Genetic_Optimizer:

    def __init__(self, mut_rate: float = 0.1, random_state: int = None):
        self.pop_size = None
        self.mut_rate = mut_rate
        self.random_state = random_state # if want start with same popluation

        # population
        self.ppl = None
        self.genes = None # Options from which our population would be created
        self.target = None
        self.target_len = None
        self.target_ls = None
        self.target_found = False
        # self.max_fitness_history = []

        self.rng = random.Random(self.random_state)
        
    def gen_ppl(self, genes, pop_size):

        genes = list(self.genes)
        # numpy for the avaliable options to generate popluation from
        ppl = np.array([np.random.choice(genes, size=self.target_len, replace=True) for _ in range(pop_size)]) 
        # ppl = np.array([np.random.choice(genes, size=target_len, replace=True) for _ in range(pop_size)])

        return ppl

                        

    def calc_fitness(self, ppl):

        '''
            loop every chromosome and check the number of characters (genes) match 
            then subtract that number from the lenght of target

            after that calculate fitness
        '''

        fnts = []

        for chrom in ppl:
            matches = 0
            for c1, c2 in zip(chrom, self.target_ls):
                if c1 == c2:
                    matches += 1

            fit = 1 / (1 + self.target_len - matches)
            fnts.append(fit)

        return fnts

    def select(self, ppl, fitness):


        new_gen_size = round(self.pop_size * .5)

        # create list with of population then store sorted version of fitness by their original indices 
        indices = sorted(range(self.pop_size), key=lambda index: fitness[index], reverse=True)
    


          # select only 50% of the popluation by loop till have
        best_ppl = self.ppl[indices[:new_gen_size]]
        return best_ppl

        
    
        
    def crossover(self, best_ppl):

        new_gen_size = self.pop_size - len(best_ppl)
        new_generation = []
        for i in range(new_gen_size):

            p1_i, p2_i = np.random.choice(len(best_ppl), size=2, replace=True)

            p1, p2 = best_ppl[p1_i], best_ppl[p2_i]


            # select one crossover point avoid first gen because does considered where
            # we use it in crossover
            crossover_point = np.random.randint(1, self.target_len)

            new_child = np.concatenate((p1[:crossover_point], p2[crossover_point:]))
            new_generation.append(new_child)


        return np.array(new_generation)



    def mut_generation(self, cr_generation):

        mut_generation = cr_generation.copy()
        # total number of genes
        total_genes = round(len(cr_generation) * len(cr_generation[0]))

        # number of genes need to mutate = total number of genes * mutation rate
        num_of_mut_genes = round(total_genes * self.mut_rate)


        # generate indices equal to the number of genes need to mutate
        gen_indices = random.sample(range(total_genes), num_of_mut_genes) 

        for index in gen_indices:
            # chromosome index
            chrom_index = index // self.target_len

            # gen index
            gen_index = index % self.target_len

            
            old_gene = mut_generation[chrom_index, gen_index]
            new_gene = random.choice([g for g in self.genes if g != old_gene])
            mut_generation[chrom_index, gen_index] = new_gene


        return mut_generation

    def fit(self, genes, target, pop_size: int = 10):



        self.target = target
        self.target_ls = list(target)
        self.target_len = len(target)
        self.genes = genes
        self.pop_size = pop_size
        # print(len(target))

        # generate the population
        self.ppl = self.gen_ppl(genes, self.pop_size)

        # fitness = self.calc_fitness(self.ppl)
        # best_ppl = self.select(self.ppl, fitness)
        # print(best_ppl)


   
                
     
        
        for i in range(10000):
            fitness = self.calc_fitness(self.ppl)
            # self.fits = fitness
            

            if i % 1000 == 0:
                print(max(fitness))

            # max_fit = max(fitness)
            # self.max_fitness_history.append(max_fit)
            # check solution found
            # for chrom in self.ppl:
            #     if chrom == self.target:
            #         print(f'Solution found chromosomes = {chrom}\ntarget = {self.target}')
            #         return

            
            best_ppl = self.select(self.ppl, fitness)
            # print(best_ppl)
            crossover_generation = self.crossover(best_ppl)
            mutat_generation = self.mut_generation(crossover_generation)
            new_generation = np.concatenate((mutat_generation, crossover_generation), axis=0)
            # print(len(new_generation))

            self.ppl = new_generation
       
            
    
```

```python
GA = Genetic_Optimizer(MUT_RATE, random_state=41)


GA.fit(GENES, TARGET, 500)
```
