```python
import random

POP_SIZE = 500
MUT_RATE = 0.1
TARGET = 'computational congitive science'

GENES = ' '
for i in range(97, 123):
    GENES += chr(i)

GENES
```

### Build the algorithm 
after finshing setting our parameter now let's build our model 


#### Model paratemer:
- `pop_size`: Pupulation size
- `mut_rate`: Mutation rate
- `radnom_state`: In case user want to start with same population
- `ppl`: Population
- `genes`: Options from which our population would be created
- `Target`: Target solution

```python
class Genetic_Optimizer:

    def __init__(self, mut_rate: float = 0.1, random_state: int = None):
        self.pop_size = None
        self.mut_rate = mut_rate
        self.random_state = random_state # if want start with same popluation

        # population
        self.ppl = None
        self.genes = None # Options from which our population would be created
        self.target = None
        self.target_len = None

        self.target_found = False

    def gen_ppl(self, genes, pop_size):

        ppl = []

        if self.random_state is not None:
            random.seed(self.random_state)

        # list for the avaliable options to generate popluation from
        for _ in range(pop_size):
            chrom = ''.join(random.choices(genes, k=self.target_len))
            # print(len(chrom))
            ppl.append(chrom)

        return ppl

                        

    def calc_fitness(self, ppl):

        '''
            loop every chromosome and check the number of characters (genes) match 
            then subtract that number from the lenght of target

            after that calculate fitness
        '''
        
        ftn = []
        
        
        for chrom in ppl:
            matches = 0
        
            for i in range(self.target_len):
                if self.target[i] == chrom[i]:
                    matches += 1
                    
            fit = 1 / (1 + abs(self.target_len - matches))
            ftn.append(fit)

        return ftn
        

    def select(self, ppl, fitness):


        new_gen_size = round(self.pop_size * .5)

        # create list with of population then store sorted version of fitness by their original indices 
        indices = sorted(range(self.pop_size), key=lambda index: fitness[index], reverse=True)

        # select only 50% of the popluation by loop till have
        best_ppl = [ppl[index] for index in indices[:new_gen_size]]
        return best_ppl


    def crossover(self, best_ppl):

        chrom_half = int(self.target_len / 2)
       
        new_generation = []
        for i in range(len(best_ppl)):
            for j in range(i + 1, len(best_ppl)):
                # print(best_ppl[i] ,best_ppl[j])
                new_chrom = best_ppl[i][:chrom_half] + best_ppl[j][chrom_half:]
                new_generation.append(new_chrom)


        return new_generation


    def mut_generation(self, new_generation):

        # total number of genes
        total_genes = round(len(new_generation) * len(new_generation[0]))

        # number of genes need to mutate = total number of genes * mutation rate
        num_of_mut_genes = round(total_genes * self.mut_rate)


        # generate indices equal to the number of genes need to mutate
        gen_indices = random.sample(range(total_genes), num_of_mut_genes)

        chrom_lists  = [list(chrom) for chrom in new_generation]
        for index in gen_indices:

            # chromosome index
            chrom_index = index // self.target_len

            # gen index
            gen_index = index % self.target_len

            chrom_lists[chrom_index][gen_index] = random.choice(self.genes)
            # # convert the string to list
            # new_chrom = list(new_generation[chrom_index])

            # # replace gen (character with random character)
            # new_chrom[gen_index] = random.choice(self.genes)

            # # covert the list to string again
            # new_chrom = ''.join(new_chrom)
            # # print(f'before: {new_generation[chrom_index]} after: {new_chrom}')

            # mutat_gen.append(new_chrom)
    
        
        mutat_gen = [''.join(chrom) for chrom in chrom_lists]
        return mutat_gen

        
        
    def fit(self, genes, target, pop_size: int = 10):

        self.target = target
        self.target_len = len(target)
        self.genes = genes
        self.pop_size = pop_size
        # print(len(target))

        # generate the population
        self.ppl = self.gen_ppl(genes, self.pop_size)

        

        
        while True:
            fitness = self.calc_fitness(self.ppl)

            # check solution found
            for chrom in self.ppl:
                if chrom == self.target:
                    print(f'Solution found chromosomes = {chrom}\ntarget = {self.target}')
                    return

            
            best_ppl = self.select(self.ppl, fitness)
            crossover_generation = self.crossover(best_ppl)
            new_generation = best_ppl + crossover_generation
            mutat_generation = self.mut_generation(new_generation)

            self.ppl = mutat_generation
       
            
    
```
